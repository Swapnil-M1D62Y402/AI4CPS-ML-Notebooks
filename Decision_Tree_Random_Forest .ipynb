## 1. Processing the Iris dataset

Firstly import all the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

Download all the iris dataset

```python
path = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
```

Assign Classnames

```python
headernames = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'Classes']
```

#Extract the data from the link as csv

```python
data = pd.read_csv(path, names=headernames)
```

```python
data.head()
```

```python
data.tail()
```

```python
data.shape
```

```python
X = data.iloc[:, :-1].values
y = data.iloc[:, 4].values
X, y
```

Split the data into Training, Testing data-Subsets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

Normalizing the data with (Mean, Var) = (0,1) with StandardScaler is a preprocessing technique in scikit-learn used for standardizing features by removing the mean and scaling to unit variance.

But this step is optional for Decision tree and Random Forest. But it is important step for KNN, SVM and Naive Bayes like classifiers

```python
from sklearn.preprocessing import StandardScaler

scale  = StandardScaler()
scale.fit(X_train)  #get the mean and variance from training set
X_train = scale.transform(X_train)
X_test = scale.transform(X_test)
```

## 2. Import the model from `sklearn.tree`

```python
from sklearn.tree import DecisionTreeClassifier

classifier = DecisionTreeClassifier()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_train)
y_pred
```

```python
acc = accuracy_score(y_train, y_pred)
print(f"Accuracy: {acc*100} %")
```

## 3. Plot the Confusion Matrix

```python
test_pred = classifier.predict(X_test)
Test_acc = accuracy_score(y_test, test_pred)
print(f"Test ACcuracy: {Test_acc*100}%")
```

```python
conf_mat = confusion_matrix(y_test, test_pred)

sns.heatmap(conf_mat,
            annot=True,
            fmt='g',
            xticklabels=['Setosa', 'Versicolor', 'Virginica'],
            yticklabels=['Setosa', 'Versicolor', 'Virginica'],)
plt.ylabel("Predicted Label", fontsize=13)
plt.xlabel("True Label", fontsize=13)
plt.title("Confusion Matrix of Decision Tree", fontsize=20)
plt.show()
```

Hence we see that for confusion matrix made on the test datasets, we are getting 100% accuracy. Out of 19 Setosa, all 19 Setosa are predicted correctly.

```python
DT_result = classification_report(y_test, test_pred)
DT_acc = accuracy_score(y_test, test_pred)
DT_result_dict = {'Classification Report': DT_result,
                  'Accuracy': DT_acc}

print(f"Decision Tree Report: {DT_result_dict}")
```

# 4. Emsemble Decision Tree

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_train)
y_pred
```

```python
test_pred = classifier.predict(X_test)
acc = accuracy_score(y_test, test_pred)
print(f"Random Forest Accuracy: {acc*100}%")
```

5. Plot confusion matrix for Random Forest

```python
conf_mat = confusion_matrix(y_test, test_pred)

sns.heatmap(conf_mat,
            annot=True,
            fmt='g',
            xticklabels=['Setosa', 'Versicolor', 'Virginica'],
            yticklabels=['Setosa', 'Versicolor', 'Virginica'])
plt.ylabel('Predicted Label', fontsize=13)
plt.xlabel('True Label', fontsize=13)
plt.title('Confusion Matrix', fontsize=20)
plt.show()
```

Here also we are getting a very high Accuracy. Such as out of 19 setosa we are correctly predicting all 19 setosas

Make classification report for Random Forest Tree
